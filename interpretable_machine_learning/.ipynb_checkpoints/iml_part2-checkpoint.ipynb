### Preparação do ambiente e do banco de dados

```r
## versão do R
R.Version()$version.string
```

```r
## pacotes
library(tidyverse) # versão 1.3.0
library(magrittr) # versão 1.5
library(h2o) # versão 3.30.0.7
library(vip) # versão 0.2.2
library(iml) # versão 0.10.0

# opções
options(repr.plot.width=7, repr.plot.height=5, repr.plot.res=200)
```

Vou ler o mesmo banco de dados utilizado na primeira parte e também modificar as variáveis da mesma maneira.

```r
## dados
df = read_csv('datasets_554905_1035602_houses_to_rent_v2.csv')

# ajusta os nomes
names(df) = c('city','area','rooms','bathrooms','parking_spaces','floor','animal','furniture',
              'hoa','rent_amount','property_tax','fire_insurance','total')
head(df)
```

```r
# cria o log do target
df = df %>%
    mutate(log_total = log(total))
```

```r
# transforma cidade em fator
df$city %<>% factor(c('São Paulo','Rio de Janeiro','Belo Horizonte','Porto Alegre','Campinas'))
```

```r
# cria a nova variavel de andar
df = df %>%
    mutate(floor_cat = case_when(
        (floor %in% c('-','1')) ~ '1 andar',
        (floor %in% c('2','3','5')) ~ '2 ao 5 andar',
        (floor %in% c('6','7','8','9','10')) ~ '6 ao 10 andar',
        (floor %in% c('11','12','13','14','15')) ~ '11 ao 15 andar',
        (floor %in% c('16','17','18','19','20')) ~ '16 ao 20 andar'
    )) %>%
    # demais serão 21+
    mutate(floor_cat = replace_na(floor_cat, '21+ andar'))

df$floor_cat %<>% factor(c('1 andar','2 ao 5 andar','6 ao 10 andar','11 ao 15 andar','16 ao 20 andar','21+ andar'))
```

```r
# cria a indicadora de animais
df = df %>% mutate(ind_animal = as.numeric(animal == 'acept'))
```

```r
# cria a indicadora de mobília
df = df %>% mutate(ind_furniture = as.numeric(furniture == 'furnished'))
```

```r
# calcula número de banheiros/número de vagas por quarto
df = df %>%
    mutate(bathrooms_ratio = bathrooms/rooms,
           parking_ratio = parking_spaces/rooms)
```

### Modelo

Para seguir as análises desse notebook, vou ajustar um *Gradient Boosting Machine* (GBM) com as variáveis do banco. A variável resposta é o logaritmo do valor total. O pacote utilizado para ajustar o modelo é o [H2O](https://docs.h2o.ai/h2o/latest-stable/h2o-docs/index.html).

```r
## inicia o h2o
h2o.init()
```

```r
# remove barras de progresso
h2o.no_progress()
```

```r
## sobe o dataset para o cluster
df_model = df %>% select(log_total, city, area, rooms, bathrooms, parking_spaces, floor_cat, ind_animal, ind_furniture)
df_model_h2o = as.h2o(df_model)

## ajusta o modelo
gbm = h2o.gbm(y = 'log_total', training_frame = df_model_h2o)
gbm
```

Não vou entrar em detalhes aqui sobre a qualidade do ajuste. Abaixo temos o gráfico de importância global das variáveis.

```r
## importância global das variáveis no modelo
vip(gbm, num_features = 8)
```

Para gerar os gráficos de interpretação das variáveis utilizarei o pacote `iml`, cuja documentação está disponível [aqui](https://cran.r-project.org/web/packages/iml/iml.pdf).

O primeiro passo é utilizar o método `new` do objeto `Predictor` para criar uma função que devolva as predições do modelo. Ele recebe como entradas o modelo e os dados. No nosso caso, a estimativa do log do valor total do aluguel é a variável resposta.

```r
## cria o preditor
predictor_gbm = Predictor$new(model = gbm, # o modelo
                              data = df_model, # os dados
                              y = 'log_total') # a resposta
```

Com o preditor construído, podemos começar a gerar os gráficos.

#### *Partial Dependence Plot (PDP)*

Todos os gráficos são construídos utilizando o objeto `FeatureEffect`, alterando somente o conjunto de variáveis avaliadas e o tipo de gráfico.

Primeiro vamos criar o PDP para a variável área. Basta criar o objeto com o método `new` e depois chamar o atributo `plot()` no objeto criado.

```r
## cria o objeto
pdp = FeatureEffect$new(predictor_gbm, feature = 'area', method = 'pdp')
pdp
```

Nos resultados podemos ver o valor estimado da *partial dependence function* em alguns pontos.

```r
# PDP de área
pdp$plot()
```

No eixo x temos os valores da variável área que encontramos no nosso banco de dados. No eixo y temos o valor estimado da *partial dependence function*. A linha, portanto, representa a relação entre a resposta e a variável área.

Temos também barras pretas logo acima do eixo x. Elas representam os valores que aparecem no banco de dados. Daí já podemos tirar a primeira conclusão: a variável área é fortemente assimétrica. Existem vários pontos em valores baixos de área, e apenas alguns valores maiores. Entretanto, quando se tem poucas observações numa faixa de valores a interpretação do PDP pode levar a conclusões errôneas.

O que fazer? Uma opção é usar o método apenas em um subset do banco de dados, excluindo os valores extremos, para tentar enxergar melhor a relação entre a variável e a resposta.

```r
# distribuição da variável área
summary(df_model$area)
```

```r
# proporção de casos em que a área é menor do que 1000
sum(df_model$area <= 1000)/nrow(df_model)
```

Como 99,9% das observações têm valores de área entre 0 e 1000, vou criar um novo objeto `Predictor` e um novo objeto `FeatureEffect` considerando apenas as observações nesse intervalo.

```r
# novo objeto PDP para area
predictor_gbm2 = Predictor$new(model = gbm,
                               data = df_model %>% filter(area <= 1000),
                               y = 'log_total')

pdp2 = FeatureEffect$new(predictor_gbm2, feature = 'area', method = 'pdp')
```

```r
# plot 2 para área
pdp2$plot()
```

Agora sim podemos ver com mais clareza tanto a curva quando as barras indicando as observações. Vamos q a maior parte das observações esta concentrada em até 250.

Analisando a curva, vemos q existe correlação positiva entre a variável e a resposta, mas ela não é linear. Quanto maior a área, maior o logaritmo do valor total, porém chega um ponto em que o log do valor total cresce muito pouco ou mesmo para de crescer.

Se eu quiser ver o PDP de uma outra variável, como por exemplo a cidade, basta mudar a referência no objeto `pdp`.

```r
# PDP de cidade
pdp$set.feature('city')
pdp$plot()
```

Como essa é uma variável categórica é gerado um gráfico de barras. Existe variação entre os valores das cidades, sendo São Paulo a cidade de maior média e Campinas a de menor. A variação não é grande e a escala do gráfico reforça essa percepção, mas é possível notar diferença clara entre os valores para São Paulo e Rio de Janeiro e as demais cidades.

Podemos também ver o resultado para a variável de segunda maior importância: o número de banheiros.

```r
# PDP do número de banheiros
pdp$set.feature('bathrooms')
pdp$plot()
```

O número de banheiros é uma variável numérica, mas assume apenas valores discretos. Também vemos aqui uma relação positiva, mas que também não é linear: a partir de 5 banheiros, não existe aumento no valor predito. É importante destacar também que não temos muitos exemplos no banco de dados de imóveis com mais de 5 banheiros e que isso pode afetar tanto a capacidade preditiva do modelo para esses pontos quanto o resultado do PDP.

#### *Individual Conditional Expectation (ICE)*

Vamos avaliar dessa vez o ICE plot da variável área. Já vou avaliar removendo as entradas maiores do que 1000.

```r
## cria o objeto
ice2 = FeatureEffect$new(predictor_gbm2, feature = 'area', method = 'ice')
ice2
```

Nesses resultados aparece o id, que é o id da observação.

```r
# ICE plot da variável área
ice2$plot()
```

Aqui já podemos ver a primeira desvantagem do ICE plot: com muitos pontos ele fica muito poluído. Mas é interessante ver que existe variabilidade na predição.

Também é possível criar o ICE plot junto com o PDP, fornecendo uma visão mais completa.

```r
## cria o objeto
ice_pdp2 = FeatureEffect$new(predictor_gbm2, feature = 'area', method = 'pdp+ice')
```

```r
# ICE plot e pdp da variavel área
ice_pdp2$plot()
```

No gráfico acima, a linha destacada é a média, ou seja, a mesma linha do PDP que fizemos antes.

No caso de variáveis categóricas é isso que acontece quando tentamos fazer um ICE plot.

```r
## cria o objeto
ice = FeatureEffect$new(predictor_gbm, feature = 'floor_cat', method = 'ice')
```

```r
# ICE plot da variável área
ice$plot()
```

A função monta um boxplot das predições. A desvantagem é que não conseguimos ver o comportamento de cada observação, mas ainda dá pra ter uma visão de variabilidade.

#### *Accumulated Local Effects (ALE)*

No caso do nosso problema, as variáveis explicativas do modelo são correlacionadas. Por exemplo, o número de quartos e a área, o número de banheiros e o número de quartos, etc. O pacote `iml` tem ferramentas para que possamos verificar a interação entre variáveis.

O gráfico abaixo mostra o nível de interação global entre as variáveis do modelo.

```r
# interações globais
interacao = Interaction$new(predictor_gbm)
plot(interacao)
```

Obs.: Na primeira chamada da função é gerada uma quantidade grande infos. No notebook isso fica um pouco poluído (foi mal!).

A força da interação (*overal interaction strength*) é calculada utilizando a *Friedman's H-statistic* e o que ela mostra é a proporção da variabilidade total do modelo é explicada pela interação entre cada variável e as demais. É possível encontrar mais informações sobre essa medida [aqui](https://christophm.github.io/interpretable-ml-book/interaction.html).

Vemos que esse valor é maior para o número de banheiros e menor para a indicadora de animais.

Podemos ver também a interação de cada variável com uma variável específica, como o número de banheiros, por exemplo.

```r
interacao_bathrooms = Interaction$new(predictor_gbm, feature = 'bathrooms')
plot(interacao_bathrooms)
```

As variáveis com maior interação com o número de banheiros são o número de vagas de garagem e o número de quartos. Nos casos em que existe correlação entre as variáveis não é recomendado usar o PDP mas sim o ALE plot.

Diferente do PDP, o ALE plot é centralizado em 0 e o valor da curva é positivo ou negativo dependendo do efeito do valor da feature na predição final.

Obs.: O ALE é o método *default* do método `new`, mas vou especificá-lo claramente por razões didáticas.

```r
## cria o objeto
ale2 = FeatureEffect$new(predictor_gbm2, feature = 'area', method = 'ale')
```

```r
# ALE plot da variável área
ale2$plot()
```

O comportamento é essencialmente o mesmo do PDP, mas o eixo y tem referências diferentes. Isso ocorre pois o ALE é uma diferença. A conclusão final é a mesma: a área tem correlação positiva com o log do valor total, mas a relação não é linear.

Vamos ver agora o ALE plot do número de banheiros, uma variável que tem correlação com outras variáveis do modelo.

```r
## cria o objeto
ale = FeatureEffect$new(predictor_gbm, feature = 'bathrooms', method = 'ale')
```

```r
# ALE plot do número de banheiros
ale$plot()
```

O ALE plot do número de banheiros é um pouco diferente do PDP. No PDP vimos alguns degraus e uma inclinação maior na curva nos valores até 5. A mudança no comportamento da curva é um reflexo da correlação entre as variáveis no modelo.

O ALE também pode ser feito para variáveis categóricas. Nesse caso ele precisa que as cateogorias tenham uma ordem (por isso importante definir como fator na criação da variável). Caso não exista ordem, o método vai ordenar as categorias utilizando similaridade em relação às demais variáveis.

```r
# ALE do andar
ale$set.feature('floor_cat')
ale$plot()
```

No ALE plot podemos ver que os modelo tem predições menores para imóveis no 1º andar. Também é possível ver que a categoria com maior efeito positivo na predição é "16 ao 20 andar".
